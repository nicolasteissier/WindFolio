```python
import urllib.request
import os

url = "https://donneespubliques.meteofrance.fr/donnees_libres/Static/listeStationsPackRadome.csv"
output_path = "../data/raw/weather/listeStations.csv"

os.makedirs(os.path.dirname(output_path), exist_ok=True)

with urllib.request.urlopen(url) as response, open(output_path, "wb") as f:
    f.write(response.read())

print(f"Downloaded to {output_path}")
```

```python
import requests, zipfile, io

url = "https://eco2mix.rte-france.com/download/eco2mix/eCO2mix_RTE_Annuel-Definitif_2012.zip"
r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))

z.extractall("../data/raw/elec/")
```
