```python
# compute returns without merging
def compute_returns(df_power, df_price):
    """
    Computes the returns based on power generation and electricity prices.

    Parameters:
    df_power (pd.DataFrame): DataFrame containing power data with a datetime index.
                            Expected format: index as datetime, column 'mwh'.
    df_price (pd.DataFrame): DataFrame containing price data with a datetime index.
                            Expected format: index as datetime, column 'price_euros_mwh'.

    Returns:
    pd.Series: Series containing the computed returns indexed by time.
    """
    # TODO: replace 'ws' with 'mwh' when power data is available
    df_returns = df_power['ws'] * df_price['price_euros_mwh']
    return df_returns

df_returns = compute_returns(df_power, df_prices)
df_returns.head()
```
